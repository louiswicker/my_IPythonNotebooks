```python
import numpy as np
import matplotlib.pyplot as plt
def func(x,y):
    return (x+y) * np.exp(-6.0*(x*x+y*y))
```

```python
xnew, ynew = np.mgrid[-1:1:70j, -1:1:70j]
z = func(xnew, ynew)
```

```python
plt.figure()
plt.pcolor(xnew, ynew, z)
plt.colorbar()
plt.title("Sparsely sampled function.")
plt.show()
```
