```python
import glob
import numpy as N
import netCDF4 as ncdf
import os
import matplotlib.pylab as P
from matplotlib.ticker import ScalarFormatter
from matplotlib.offsetbox import AnchoredText
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.ticker as ticker

from plot_cookbook import *

_time_units    = "seconds since 2011-05-24 19:40:00"
_calendar      = 'standard'

file = "adv_1min.nc"

f = ncdf.Dataset(file)

print(f)
```

```python
RF1 = N.ma.masked_greater_equal(f.variables['REF1'][...], 998.9)
VR   = N.ma.masked_greater_equal(f.variables['VR'][...], 998.9)
UT   = N.ma.masked_greater_equal(f.variables['UT'][...], 998.9)
VT   = N.ma.masked_greater_equal(f.variables['VT'][...], 998.9)
```

```python
fig, axes =  P.subplots(2, 2, figsize=(24,24))
zoom=[0.2,0.7,0.3,0.8]
plot_contour(VR[7], title = "VR", ax = axes[0][0], scale=1, cint=0., zoom=zoom)
plot_contour(RF1, title = "REF", ax = axes[0][1], scale=1, zoom=zoom, plot_defaults='REF')
plot_contour(UT, title = "U", ax = axes[1][0], scale=1, ctable='bwr', cint=0., zoom=zoom)
plot_contour(VT, title = "V", ax = axes[1][1], scale=1, ctable='bwr', cint=0., zoom=zoom)
P.show()
```
