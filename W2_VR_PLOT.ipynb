```python
def plotW2vr(filename = None, elevation = None, output = None, directory = "./", show = True):
    from netcdftime import utime
    from mpl_toolkits.axes_grid.inset_locator import inset_axes
    from mpl_toolkits.basemap import Basemap
#   from pyesviewer import ctables
    
    sec_utime = utime("seconds since 1970-01-01 00:00:00")
    
    f = ncdf.Dataset(filename, "r")
    vr = f.variables['data'][:]
    vr = N.ma.masked_array(vr, mask=(vr == f.data_below_snr))
    
    el = f.variables['elevation_angle'][:]
    lat = f.variables['latitude'][:]   
    lon = f.variables['longitude'][:]
    timestamp = f.variables['time_since_1970'][:]    

    if elevation <= 0:
        k = -elevation
    else:
        kel = el[el > elevation]
        if kel.size > 0:  
            k = kel[0] 
        else:
            k = 0
            
    vrslice = vr[k]

    fig_y_size = 8.0 * (lat.max() - lat.min()) / (lon.max()-lon.min())
    fig = P.figure(figsize=(12,12))
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')

    map = Basemap(projection='lcc', llcrnrlon=lon[0][0],llcrnrlat=lat[0][0],urcrnrlon=lon[-1][-1],urcrnrlat=lat[-1][-1], lat_1=30.,lat_2=50.,lon_0=-95., area_thresh=1.,suppress_ticks=True)
    
    clevels = N.arange(-30,35,5)
    plot = ax.contourf(lon, lat, vrslice, clevels)
    axins = inset_axes(ax, width="5%", height="35%", loc=3)
    locator=axins.get_axes_locator()
    locator.set_bbox_to_anchor((1.05, 0.05, 1, 1), ax.transAxes)
    locator.borderpad = 0.
    cbar = fig.colorbar(plot, format="%d", cax=axins)
    cbar.ax.set_title('( %s )\n' % 'Vr m/s', ha='left')

    ax.set_title("Time:  %s       Vr SuperObs at %4.2f deg" % (sec_utime.num2date(timestamp[k]), el[k]), size='xx-large')
    ax.set_aspect('equal')
    ax.set_xlabel('Lon (km)')
    ax.set_ylabel('Lat (km)')
    
    try:
        shapelist = os.getenv("PYESVIEWER_SHAPEFILES").split(":")

        if len(shapelist) > 0:

            mapaxes1 = inset_axes(ax,width="100%",height="100%",loc=3)
            mapaxes1.set_axis_off()
            locator1 = mapaxes1.get_axes_locator()
            locator1.set_bbox_to_anchor((0,0,1,1),ax.transAxes)
            locator1.borderpad = 0.0

            for item in shapelist:
                items = item.split(",")
                shapefile  = items[0]
                color      = items[1]
                line_width = items[2]
#                print shapefile, color, line_width

                map.readshapefile(shapefile,'GIS_INFO',drawbounds=True,linewidth=line_width,color=color,ax=mapaxes1)

    except OSError:
        print "GIS_PLOT:  NO SHAPEFILE ENV VARIABLE FOUND "
        pass

    output_filename = "%s_%4.2f.png" % (os.path.join(directory,str.replace(sec_utime.num2date(timestamp[k]).isoformat(),"T","_")), elevation)
    if output != None:
        print "\n Saving file %s" % (output_filename)
        P.savefig(output_filename, format="png")
        P.close()
    if show == True:
        P.show()
        P.close()
        
```
